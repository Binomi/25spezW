```python
import math
R=8.314
```

```python
def f(x):
    return x**3/(math.exp(x)-1)

def c_m(T,debye):
    integral=0.0
    t=0.0
    dt=0.01
    while t<=T/debye:
        t+=dt
        integral+=dt*f(t)
    return 9.0*R*((T/debye)**3)*(4*integral-((debye/T)**4)/(math.exp(debye/T)-1))
```

```python
debye=440.0
T=1.0
while T<400.0:
    print c_m(T,debye)
    T+=1.0
```
